```python
import random
import matplotlib.pyplot as plt
```

```python

def draw_ifs_fractal(rules, probabilities, n_points=100000):
    x, y = 0.0, 0.0
    xs, ys = [], []
    
    for _ in range(n_points):
        r = random.random()
        total = 0
        for i, p in enumerate(probabilities):
            total += p
            if r <= total:
                a, b, c, d, e, f = rules[i]
                x, y = a*x + b*y + e, c*x + d*y + f
                xs.append(x)
                ys.append(y)
                break
    
    plt.figure(figsize=(10, 10))
    plt.scatter(xs, ys, s=0.1, c='green', alpha=0.6, marker='.')
    plt.axis('off')
    plt.gca().set_aspect('equal')
    plt.show()
```

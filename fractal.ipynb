Визначити фрактальну розмірність фрактала на площині, що складається з
точок (x; y), де x ∈ [0; 1], y ∈ [0; 1], причому в десятковому представленні
чисел x і y відсутні цифри 3 та 7. Розробити програмне забезпечення для
побудови даного фрактала.

a = [...] len(a) = 8

D = log N(l)/ log l

=> D = log(64)/log 10 = log(8^2)/ log 10 = 2 log(8)/ log 10

D = 2 lg(8) = 2 lg(2^3) = 6 lg(2) = 1.80618 

```python
import matplotlib.pyplot as plt
import itertools
```

```python
def fractal(depth):
    
    a = [str(d) for d in range(10) if d not in (3, 7)]

    x_list = []
    y_list = []
    

    for pairs in itertools.product([(dx, dy) for dx in a for dy in a], repeat=depth):
        
        x_acc = ''.join(p[0] for p in pairs)
        y_acc = ''.join(p[1] for p in pairs)
        x = int(x_acc) / 10**depth
        y = int(y_acc) / 10**depth
        x_list.append(x)
        y_list.append(y)
    
    return x_list, y_list


x, y = fractal(2)


plt.figure(figsize=(9, 9))
plt.scatter(x, y, s=0.5, color='black')

plt.xlabel("x")
plt.ylabel("y")
plt.grid(True, linestyle='--', alpha=0.5)

plt.xlim(0, 1)
plt.ylim(0, 1)

plt.tight_layout()
plt.show()
```

```python
x, y = fractal(3)


plt.figure(figsize=(9, 9))
plt.scatter(x, y, s=0.5, color='black')

plt.xlabel("x")
plt.ylabel("y")
plt.grid(True, linestyle='--', alpha=0.5)

plt.xlim(0, 1)
plt.ylim(0, 1)

plt.tight_layout()
plt.show()
```
